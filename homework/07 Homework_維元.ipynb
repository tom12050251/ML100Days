### 1. 請比較對一個 1000 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。




```python
import numpy as np
```

```python
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
```

```python
%timeit -n 10 a = 2
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
print(Z[0:]+1)
```

```python
%timeit -n 10 a = 2
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
for row in Z:
    print(row+1)
```

```python
%timeit -n 10 a = 2
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
for row in Z:
    for d in row:
        print(d + 1)
```

### 2. 如何從一個陣列中，找出出現頻率最高的數值與位置？


```python
Z = np.random.randint(0,10,50)
print(Z)
```

```python
counts = np.bincount(Z)
Zmod=np.argmax(counts)
print(Zmod)
```

```python
np.where(Z == Zmod)
```

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。

```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))


b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))

for row in b:
    print(row)


c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
```

```python
def tolist(iterable):
    if type(iterable) != np.ndarray:
        return iterable
    
    newlist=[]
    for row in iterable:
        newlist.append(tolist(row))
    return newlist

print(tolist(b))
print(tolist(c))
```
